```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from merlin import post
import itertools

path = '/Users/bnowicki/Documents/Scratch/analysis_3/analysis/main_dir/'
simpost = post.Simulation_Post_Analysis('text1x', path)

df = simpost.populate_table()
```

```python
lines=["H1_6562.80A","O1_1304.86A","O1_6300.30A","O2_3728.80A","O2_3726.10A",
       "O3_1660.81A","O3_1666.15A","O3_4363.21A","O3_4958.91A","O3_5006.84A", 
       "He2_1640.41A","C2_1335.66A","C3_1906.68A","C3_1908.73A","C4_1549.00A",
       "Mg2_2795.53A","Mg2_2802.71A","Ne3_3868.76A","Ne3_3967.47A","N5_1238.82A",
       "N5_1242.80A","N4_1486.50A","N3_1749.67A","S2_6716.44A","S2_6730.82A"]

column_list = []

for line in lines:
    field = ('gas', f'luminosity_{line}')
    column = str(field) + '_agg'
    column_list.append(column)

print(column_list)

plt.figure(figsize=(8, 6))

for i, column in enumerate(column_list):
    plt.plot(df['current_redshift'], np.log10(df[column]), linestyle='-', label=lines[i])
plt.xlabel('Redshift')
plt.ylabel('Luminosities')
#plt.title(title)
plt.grid(True)
ax = plt.gca()
ax.invert_xaxis()
plt.tight_layout()
plt.legend()
#plt.savefig("plot.png")
plt.show()
```

```python
# Define color and linestyle cycles
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
linestyles = ['-', '--', '-.', ':']

# Create a combined (color, linestyle) cycle
style_cycler = itertools.cycle(itertools.product(colors, linestyles))

plt.figure(figsize=(10, 6))

for i, column in enumerate(column_list):
    if column in df.columns:
        y_vals = df[column].replace(0, np.nan)
        color, linestyle = next(style_cycler)
        plt.plot(df['current_redshift'], np.log10(y_vals), linestyle=linestyle,
                 color=color, label=lines[i])

plt.xlabel('Redshift')
plt.ylabel('Luminosities')
plt.grid(True)
ax = plt.gca()
ax.invert_xaxis()
plt.tight_layout()

# Place legend to the right
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)

plt.show()
```

```python
from collections import defaultdict

# Group lines by element (e.g., 'O3', 'H1', etc.)
groups = defaultdict(list)
for line in lines:
    prefix = line.split('_')[0]
    groups[prefix].append(line)

# Set up color and linestyle cycler
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
linestyles = ['-', '--', '-.', ':']
style_cycler = itertools.cycle(itertools.product(colors, linestyles))

plt.figure(figsize=(10, 6))

for group, group_lines in groups.items():
    color, linestyle = next(style_cycler)
    for i, line in enumerate(group_lines):
        column = f"('gas', 'luminosity_{line}')_agg"
        if column in df.columns:
            y_vals = df[column].replace(0, np.nan)
            # Only label the first line of the group
            label = group if i == 0 else None
            plt.plot(df['current_redshift'], np.log10(y_vals),
                     color=color, linestyle=linestyle, label=label)

plt.xlabel('Redshift')
plt.ylabel('Luminosities')
plt.grid(True)
ax = plt.gca()
ax.invert_xaxis()
plt.tight_layout()

# Show unique element legend to the right
plt.legend(title='Element', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.show()
```
